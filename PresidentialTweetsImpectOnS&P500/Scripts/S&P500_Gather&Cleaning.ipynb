```python
# Prerequisites
# pip install yfinance

import pandas as pd
import yfinance as yf
```

```python
sp500 = yf.download("^GSPC", start="2017-01-01", end="2021-01-01")
print(sp500.head())
```

```python
sp500_cleaned = sp500.dropna()
sp500_cleaned = sp500_cleaned[['Close']]
sp500_cleaned = sp500_cleaned.reset_index()
sp500_cleaned.to_csv('SP500_IndexCloses.csv', index=False)
```

```python
sp500_tickers = pd.read_csv('/Users/keremsmacbook/Desktop/42/gitHub/Presidential_Tweets_Impacts_SP500/PresidentialTweetsImpectOnS&P500/Data/Raw/sp500_tickers.csv')

start_date = "2010-01-01"
end_date = "2023-12-31"

all_data = pd.DataFrame()
for ticker in sp500_tickers['0']:  
    data = yf.download(ticker, start=start_date, end=end_date)

    if 'Close' in data.columns:
        all_data[ticker] = data['Close']

all_data.to_csv('SP500_Closing_Prices.csv')
```
